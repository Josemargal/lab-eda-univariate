Generar tabla de frecuencias para la categoría de productos:

```python
# Importar bibliotecas
import pandas as pd

# Cargar el archivo CSV descargado
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')

# Mostrar las primeras filas del dataset
print(df.head())

# Generar tabla de frecuencias
freq_table = df['category'].value_counts()
print(freq_table.head(5))  # Mostrar las 5 categorías más frecuentes
```

Visualización de la distribución de productos por categorías:

```python
# Visualización con un gráfico de barras
import matplotlib.pyplot as plt
import seaborn as sns

top_categories = freq_table.head(10)  # Las 10 categorías principales
plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title('Distribución de productos por categoría')
plt.xlabel('Cantidad de productos')
plt.ylabel('Categoría')
plt.show()
```

Visualización de proporciones de categorías principales con gráfico circular:

```python
# Gráfico de torta para las 5 categorías principales
top_5_categories = freq_table.head(5)
plt.figure(figsize=(7,7))
plt.pie(top_5_categories.values, labels=top_5_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Proporciones de las categorías principales')
plt.show()
```

Parte 2: Análisis de precios de productos
Medidas de centralidad (media, mediana, moda):

```python
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
print(f'Media: {mean_price}, Mediana: {median_price}, Moda: {mode_price}')
```

Medidas de dispersión (varianza, desviación estándar, rango, rango intercuartílico):

```python
price_variance = df['price'].var()
price_std_dev = df['price'].std()
price_range = df['price'].max() - df['price'].min()
price_iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)
print(f'Varianza: {price_variance}, Desviación Estándar: {price_std_dev}, Rango: {price_range}, IQR: {price_iqr}')
```

Histograma de la distribución de precios:

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribución de precios de productos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

Boxplot para identificar productos con precios significativamente altos:

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['price'])
plt.title('Boxplot de precios de productos')
plt.show()
```

Parte 3: Análisis de valoraciones de productos
Medidas de centralidad (media, mediana, moda):

```python
mean_rating = df['reviews'].mean()
median_rating = df['reviews'].median()
mode_rating = df['reviews'].mode()[0]
print(f'Media: {mean_rating}, Mediana: {median_rating}, Moda: {mode_rating}')
```

Medidas de dispersión (varianza, desviación estándar, rango intercuartílico):

```python
rating_variance = df['reviews'].var()
rating_std_dev = df['reviews'].std()
rating_iqr = df['reviews'].quantile(0.75) - df['reviews'].quantile(0.25)
print(f'Varianza: {rating_variance}, Desviación Estándar: {rating_std_dev}, IQR: {rating_iqr}')
```

Cálculo de asimetría (skewness) y curtosis:

```python
skewness = df['reviews'].skew()
kurtosis = df['reviews'].kurt()
print(f'Asimetría: {skewness}, Curtosis: {kurtosis}')
```

Histograma de la distribución de valoraciones:

```python
plt.figure(figsize=(10,6))
sns.histplot(df['reviews'], bins=10, kde=True)
plt.title('Distribución de valoraciones de productos')
plt.xlabel('Valoración')
plt.ylabel('Frecuencia')
plt.show()
```
